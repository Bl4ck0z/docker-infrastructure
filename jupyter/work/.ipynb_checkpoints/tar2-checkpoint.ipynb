```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Configuración para mejores gráficos en Jupyter
plt.style.use('default')
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 10
```

```python
# =========================================
# 1. IMPORTACIÓN DE DATOS
# =========================================
print("\n1. IMPORTACIÓN DE DATOS")
print("-" * 40)

# Cargar dataset desde archivo CSV
df = pd.read_csv('netflix_dataset_ficticio.csv')

# Información básica del dataset
print(f"Dataset cargado exitosamente")
print(f"Dimensiones: {df.shape[0]} filas × {df.shape[1]} columnas")
print(f"Fecha de análisis: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")

print(f"\nEstructura inicial del dataset:")
display(df.head())

# Evaluación preliminar de la calidad de datos
print(f"\nEvaluación inicial de calidad:")
if df.isna().any().any():
    print("ESTADO: Requiere limpieza - valores nulos detectados")
else:
    print("ESTADO: Dataset limpio - sin valores nulos detectados")
```

```python
# =========================================
# 2. EXPLORACIÓN INICIAL
# =========================================
print(f"\n2. EXPLORACIÓN INICIAL DE DATOS (EDA)")
print("-" * 40)

# Dimensiones del dataset
print(f"Registros totales: {df.shape[0]}")
print(f"Variables/columnas: {df.shape[1]}")

# Análisis de valores nulos
nulls_summary = df.isnull().sum().sort_values(ascending=False)
nulls_df = pd.DataFrame({
    'Columna': nulls_summary.index,
    'Valores_Nulos': nulls_summary.values,
    'Porcentaje': (nulls_summary.values / len(df) * 100).round(2)
})

print(f"\nAnálisis de completitud de datos:")
display(nulls_df[nulls_df['Valores_Nulos'] > 0])

# Análisis de tipos de datos
print(f"\nEstructura de tipos de datos:")
dtypes_df = pd.DataFrame({
    'Columna': df.dtypes.index,
    'Tipo_Datos': df.dtypes.astype(str).values
})
display(dtypes_df)

# Estadísticas descriptivas para variables numéricas
numeric_cols = df.select_dtypes(include=[np.number]).columns
if len(numeric_cols) > 0:
    print(f"\nEstadísticas descriptivas (variables numéricas):")
    display(df[numeric_cols].describe())

# Identificación de problemas de calidad de datos
print(f"\nDiagnóstico de calidad de datos:")
quality_issues = []

if df.isnull().sum().sum() > 0:
    quality_issues.append("• Presencia de valores nulos requiere estrategia de imputación")

if 'date_added' in df.columns and not pd.api.types.is_datetime64_any_dtype(df['date_added']):
    quality_issues.append("• Variable 'date_added' requiere conversión a tipo datetime")

if 'release_year' in df.columns and not pd.api.types.is_numeric_dtype(df['release_year']):
    quality_issues.append("• Variable 'release_year' requiere conversión a tipo numérico")

if quality_issues:
    for issue in quality_issues:
        print(issue)
else:
    print("No se detectaron problemas críticos de calidad")
```

```python
# =========================================
# 3. TRANSFORMACIÓN Y LIMPIEZA DE DATOS
# =========================================
print(f"\n3. TRANSFORMACIÓN Y LIMPIEZA DE DATOS")
print("-" * 40)

# Crear copia del dataset para transformaciones
df_clean = df.copy()

# Estrategia de imputación para variables categóricas
categorical_columns = ['director', 'cast', 'country', 'rating', 'listed_in', 'description', 'title']
for col in categorical_columns:
    if col in df_clean.columns:
        df_clean[col] = df_clean[col].fillna('Desconocido')

print("Imputación completada: Variables categóricas rellenadas con 'Desconocido'")

# Conversión de tipos de datos
if 'date_added' in df_clean.columns:
    df_clean['date_added'] = pd.to_datetime(df_clean['date_added'], errors='coerce')
    print("Conversión exitosa: 'date_added' transformado a tipo datetime")

if 'release_year' in df_clean.columns:
    df_clean['release_year'] = pd.to_numeric(df_clean['release_year'], errors='coerce').astype('Int64')
    print("Conversión exitosa: 'release_year' transformado a tipo entero")

# Feature engineering: creación de variable duration_minutes
if 'duration' in df_clean.columns:
    def extract_duration_minutes(duration_str):
        """
        Extrae la duración en minutos para películas.
        Para series devuelve NaN ya que se mide en temporadas/episodios.
        
        Args:
            duration_str (str): String con duración (ej: "95 min", "2 Seasons")
            
        Returns:
            float: Duración en minutos o NaN
        """
        if pd.isna(duration_str):
            return np.nan
        
        duration_str = str(duration_str).strip().lower()
        
        if 'min' in duration_str:
            # Extraer componente numérico para películas
            numeric_part = ''.join([c for c in duration_str if c.isdigit()])
            return float(numeric_part) if numeric_part else np.nan
        else:
            # Series se mantienen como NaN para este análisis
            return np.nan
    
    df_clean['duration_minutes'] = df_clean['duration'].apply(extract_duration_minutes)
    print("Feature engineering: Variable 'duration_minutes' creada para análisis de películas")

# Validación post-transformación
print(f"\nResumen de transformaciones aplicadas:")
display(df_clean.head())

print(f"\nMétricas de transformación:")
print(f"• Registros procesados: {len(df_clean)}")
print(f"• Variables originales: {len(df.columns)}")
print(f"• Variables post-transformación: {len(df_clean.columns)}")
print(f"• Nuevas variables creadas: {len(df_clean.columns) - len(df.columns)}")
```

```python
# =========================================
# 4. MANIPULACIÓN Y FILTRADO DE DATOS
# =========================================
print(f"\n4. MANIPULACIÓN Y FILTRADO DE DATOS")
print("-" * 40)

# Filtrado temporal: contenido posterior a 2010
if 'release_year' in df_clean.columns:
    after_2010 = df_clean[df_clean['release_year'] > 2010].copy()
    filter_efficiency = len(after_2010) / len(df_clean) * 100
    print(f"Filtrado temporal aplicado: {len(after_2010)} registros ({filter_efficiency:.1f}% del dataset)")
else:
    after_2010 = df_clean.copy()
    print("ADVERTENCIA: Filtrado temporal no aplicado - variable 'release_year' no disponible")

# Análisis agregado por país (manejo de valores múltiples)
if 'country' in after_2010.columns:
    # Normalización de datos: expansión de países múltiples por título
    countries_expanded = after_2010.assign(
        country=after_2010['country'].str.split(',')
    ).explode('country')
    
    # Limpieza de espacios en blanco
    countries_expanded['country'] = countries_expanded['country'].str.strip()
    
    # Agregación: conteo de títulos por país
    country_counts = countries_expanded.groupby('country')['title'].count().sort_values(ascending=False)
    top10_countries = country_counts.head(10)
    
    print(f"\nAnálisis geográfico - Top 10 países (contenido >2010):")
    top10_df = pd.DataFrame({
        'País': top10_countries.index,
        'Cantidad_Titulos': top10_countries.values,
        'Participacion_Porcentual': (top10_countries.values / top10_countries.sum() * 100).round(2)
    })
    display(top10_df)
    
    # Insight clave
    market_leader = top10_countries.index[0]
    market_share = top10_countries.iloc[0] / top10_countries.sum() * 100
    print(f"INSIGHT: {market_leader} lidera con {market_share:.1f}% de participación en el top 10")
```

```python
# =========================================
# 5. ANÁLISIS AVANZADO Y SEGMENTACIÓN
# =========================================
print(f"\n5. ANÁLISIS AVANZADO Y SEGMENTACIÓN")
print("-" * 40)

# Análisis de géneros (contenido >2010)
if 'listed_in' in after_2010.columns:
    # Normalización: expansión de géneros múltiples
    genres_expanded = after_2010.assign(
        listed_in=after_2010['listed_in'].str.split(',')
    ).explode('listed_in')
    genres_expanded['listed_in'] = genres_expanded['listed_in'].str.strip()
    
    # Análisis de frecuencias
    genre_counts = genres_expanded['listed_in'].value_counts()
    dominant_genre = genre_counts.index[0] if not genre_counts.empty else 'No determinado'
    
    print(f"Análisis de géneros cinematográficos (>2010):")
    print(f"Género dominante: {dominant_genre}")
    
    print(f"\nDistribución Top 5 géneros:")
    genre_distribution = pd.DataFrame({
        'Genero': genre_counts.head().index,
        'Frecuencia': genre_counts.head().values,
        'Porcentaje': (genre_counts.head().values / genre_counts.sum() * 100).round(2)
    })
    display(genre_distribution)

# Análisis de productividad por director
if 'director' in df_clean.columns:
    # Filtrar directores conocidos
    known_directors = df_clean[df_clean['director'] != 'Desconocido']['director'].value_counts()
    
    if not known_directors.empty:
        most_productive_director = known_directors.index[0]
        productivity_count = known_directors.iloc[0]
        
        # Portfolio del director más productivo
        director_portfolio = df_clean[df_clean['director'] == most_productive_director]['title'].tolist()
        
        print(f"\nAnálisis de productividad - Directores:")
        print(f"Director más productivo: {most_productive_director}")
        print(f"Cantidad de títulos: {productivity_count}")
        
        print(f"\nPortfolio de obras (muestra primeros 10 títulos):")
        for i, title in enumerate(director_portfolio[:10], 1):
            print(f"   {i:2d}. {title}")
        
        if len(director_portfolio) > 10:
            print(f"   ... y {len(director_portfolio)-10} títulos adicionales")

# Análisis comparativo de duración por mercado geográfico
if 'duration_minutes' in df_clean.columns and 'type' in df_clean.columns:
    # Segmentación: solo películas con duración válida
    movies_segment = df_clean[
        (df_clean['type'].str.lower() == 'movie') & 
        (df_clean['duration_minutes'].notna())
    ].copy()
    
    if not movies_segment.empty:
        # Expansión geográfica para películas
        movies_geo_expanded = movies_segment.assign(
            country=movies_segment['country'].str.split(',')
        ).explode('country')
        movies_geo_expanded['country'] = movies_geo_expanded['country'].str.strip()
        
        # Análisis estadístico por país
        duration_stats_by_country = movies_geo_expanded.groupby('country')['duration_minutes'].agg([
            'mean', 'median', 'std', 'count'
        ]).round(2).sort_values('mean', ascending=False)
        
        # Filtrar países con muestra significativa (>= 3 películas)
        significant_markets = duration_stats_by_country[duration_stats_by_country['count'] >= 3]
        
        print(f"\nAnálisis comparativo - Duración de películas por mercado:")
        print(f"Mercados con películas de MAYOR duración promedio:")
        display(significant_markets.head().rename(columns={
            'mean': 'Promedio_min', 'median': 'Mediana_min', 
            'std': 'Desv_Std', 'count': 'N_Peliculas'
        }))
        
        print(f"Mercados con películas de MENOR duración promedio:")
        display(significant_markets.tail().rename(columns={
            'mean': 'Promedio_min', 'median': 'Mediana_min', 
            'std': 'Desv_Std', 'count': 'N_Peliculas'
        }))
```

```python
# =========================================
# 6. VISUALIZACIÓN DE DATOS
# =========================================
print(f"\n6. VISUALIZACIÓN DE DATOS")
print("-" * 40)

# Gráfico 1: Análisis geográfico - Top 10 países
if 'country' in after_2010.columns and not top10_countries.empty:
    plt.figure(figsize=(12, 6))
    bars = plt.bar(range(len(top10_countries)), top10_countries.values, color='steelblue', alpha=0.8)
    
    plt.title('Distribución Geográfica del Contenido Netflix (Post-2010)', 
              fontsize=14, fontweight='bold', pad=20)
    plt.xlabel('País', fontsize=12)
    plt.ylabel('Número de Títulos', fontsize=12)
    plt.xticks(range(len(top10_countries)), top10_countries.index, rotation=45, ha='right')
    
    # Anotaciones de valores
    for i, bar in enumerate(bars):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 1, 
                str(top10_countries.iloc[i]), ha='center', va='bottom', fontweight='bold')
    
    plt.grid(True, axis='y', alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    print("Gráfico 1: Distribución geográfica del contenido generado")

# Gráfico 2: Análisis de géneros - Distribución porcentual
if 'listed_in' in after_2010.columns and not genre_counts.empty:
    plt.figure(figsize=(10, 8))
    
    # Preparación de datos: Top 8 géneros + categoría "Otros"
    top_genres = genre_counts.head(8)
    others_sum = genre_counts[8:].sum()
    
    if others_sum > 0:
        plot_data = list(top_genres.values) + [others_sum]
        plot_labels = list(top_genres.index) + ['Otros']
    else:
        plot_data = top_genres.values
        plot_labels = top_genres.index
    
    # Configuración de colores profesionales
    colors = plt.cm.Set3(np.linspace(0, 1, len(plot_data)))
    
    wedges, texts, autotexts = plt.pie(plot_data, labels=plot_labels, autopct='%1.1f%%', 
                                      startangle=90, colors=colors, pctdistance=0.85)
    
    plt.title('Distribución de Géneros Cinematográficos (Post-2010)', 
              fontsize=14, fontweight='bold', pad=20)
    
    # Mejorar legibilidad de porcentajes
    for autotext in autotexts:
        autotext.set_color('white')
        autotext.set_fontweight('bold')
    
    plt.axis('equal')
    plt.tight_layout()
    plt.show()
    
    print("Gráfico 2: Distribución de géneros cinematográficos generado")

# Gráfico 3: Análisis comparativo de duración por mercado
if 'duration_minutes' in df_clean.columns and not movies_geo_expanded.empty:
    # Selección de mercados más significativos para análisis visual
    top_movie_markets = movies_geo_expanded['country'].value_counts().head(5)
    
    plt.figure(figsize=(12, 8))
    
    # Preparación de datos para boxplot
    boxplot_data = []
    boxplot_labels = []
    
    for country in top_movie_markets.index:
        country_durations = movies_geo_expanded[
            movies_geo_expanded['country'] == country
        ]['duration_minutes'].dropna()
        
        if len(country_durations) > 0:
            boxplot_data.append(country_durations)
            boxplot_labels.append(f"{country}\n(n={len(country_durations)})")
    
    if boxplot_data:
        box_plot = plt.boxplot(boxplot_data, labels=boxplot_labels, patch_artist=True)
        
        # Configuración estética
        colors = ['lightblue', 'lightgreen', 'lightcoral', 'lightyellow', 'lightpink']
        for patch, color in zip(box_plot['boxes'], colors[:len(box_plot['boxes'])]):
            patch.set_facecolor(color)
            patch.set_alpha(0.7)
        
        plt.title('Análisis Comparativo: Distribución de Duración de Películas por Mercado\n(Top 5 mercados por volumen)', 
                  fontsize=14, fontweight='bold', pad=20)
        plt.xlabel('Mercado Geográfico', fontsize=12)
        plt.ylabel('Duración (minutos)', fontsize=12)
        plt.grid(True, axis='y', alpha=0.3)
        plt.tight_layout()
        plt.show()
        
        print("Gráfico 3: Análisis comparativo de duración por mercado generado")
```

```python
# =========================================
# 7. SÍNTESIS DE RESULTADOS Y CONCLUSIONES
# =========================================
print(f"\n7. SÍNTESIS DE RESULTADOS Y CONCLUSIONES")
print("=" * 60)

print(f"HALLAZGOS PRINCIPALES:")
print(f"-" * 30)

# Conclusiones geográficas
if 'country' in after_2010.columns and not top10_countries.empty:
    print(f"ANÁLISIS GEOGRÁFICO:")
    print(f"  • Mercado dominante: {top10_countries.index[0]} ({top10_countries.iloc[0]} títulos)")
    print(f"  • Segundo mercado: {top10_countries.index[1]} ({top10_countries.iloc[1]} títulos)")
    print(f"  • Tercer mercado: {top10_countries.index[2]} ({top10_countries.iloc[2]} títulos)")
    
    concentration_ratio = top10_countries.head(3).sum() / top10_countries.sum() * 100
    print(f"  • Concentración Top-3: {concentration_ratio:.1f}% del contenido analizado")

# Conclusiones de género
if 'listed_in' in after_2010.columns:
    print(f"\nANÁLISIS DE GÉNEROS:")
    print(f"  • Género predominante: {dominant_genre}")
    
    if not genre_counts.empty:
        genre_concentration = genre_counts.iloc[0] / genre_counts.sum() * 100
        print(f"  • Concentración del género líder: {genre_concentration:.1f}%")

# Conclusiones de productividad
if 'director' in df_clean.columns and not known_directors.empty:
    print(f"\nANÁLISIS DE PRODUCTIVIDAD:")
    print(f"  • Director más prolífico: {most_productive_director} ({productivity_count} títulos)")

# Conclusiones de duración
if 'duration_minutes' in df_clean.columns and not significant_markets.empty:
    longest_market = significant_markets.index[0]
    longest_duration = significant_markets.iloc[0]['mean']
    shortest_market = significant_markets.index[-1]
    shortest_duration = significant_markets.iloc[-1]['mean']
    
    print(f"\nANÁLISIS DE DURACIÓN:")
    print(f"  • Películas más extensas: {longest_market} ({longest_duration:.1f} min promedio)")
    print(f"  • Películas más concisas: {shortest_market} ({shortest_duration:.1f} min promedio)")
    
    duration_range = longest_duration - shortest_duration
    print(f"  • Rango de variación: {duration_range:.1f} minutos entre mercados")

# Insights estratégicos
print(f"\nINSIGHTS ESTRATÉGICOS:")
print(f"-" * 30)
print(f"• Concentración geográfica significativa sugiere oportunidades de diversificación")
print(f"• La distribución de géneros refleja preferencias del mercado global")
print(f"• Las diferencias en duración por país indican variaciones culturales en storytelling")
print(f"• El análisis post-2010 captura tendencias del streaming moderno")

# Métricas finales del análisis
print(f"\nMÉTRICAS DEL ANÁLISIS:")
print(f"-" * 30)
print(f"• Registros analizados: {len(df_clean):,}")
print(f"• Período temporal: Contenido posterior a 2010")
print(f"• Mercados geográficos identificados: {len(country_counts) if 'country_counts' in locals() else 'N/A'}")
print(f"• Géneros catalogados: {len(genre_counts) if 'genre_counts' in locals() else 'N/A'}")
print(f"• Visualizaciones generadas: 3 gráficos analíticos")

print("=" * 60)
print("ANÁLISIS COMPLETADO - REPORTE GENERADO EN JUPYTER NOTEBOOK")
print("=" * 60)
```
